```python
import pandas as pd
df2= {'Team':['Riders','Riders','Devils','Devils','Kings','Kings','Kings','Kings','Riders','Royals','Royals','Royals'],
            'Rank':[1,2,2,3,3,4,1,1,2,4,1,2],
            'Year':[2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
            'Points':[876,789,863,673,741,812,756,788,694,701,884,690],
            
           }
df=pd.DataFrame(df2)
df
```

```python
# Q1. Har team ka year-wise average points nikaalo aur batao kaunsi team me sabse zyada consistency hai.
df.pivot_table(
    index='Year',
    columns='Team',
    values='Points',
    aggfunc='mean',
    fill_value=0,
    margins=True
    
)
```

```python
# Q2. Kaunsi team ne Rank 1 sabse zyada baar achieve kiya hai?
df[df['Rank']==1]['Team'].value_counts()
```

```python
# Q3. Har year ke liye highest points wali team ka naam aur points nikaalo.
df.loc[df.groupby('Year')['Points'].idxmax()]
```

```python
# Q4. Har team ke liye best year (maximum points) identify karo.
df.loc[df.groupby('Team')['Points'].idxmax()]
```

```python
# Q5. 2014 aur 2015 ke beech kis team ka performance sabse zyada improve hua (points ke basis par)?
df_14_15 = df[df['Year'].isin([2014, 2015])]
df_14_15
```

```python
pivot=df_14_15.pivot(
     index='Team',
     columns='Year',
     values='Points',
     
 )
pivot
```

```python
pivot['Improvement']=pivot[2015]-pivot[2014]
pivot
```

```python
pivot['Improvement'].idxmax()
```

```python
# Q6. jo kam se kam 2 alag-alag years me Rank 1 par rahi ho.
df[df['Rank']==1]['Team'].value_counts()[lambda x: x>=2]
```

```python
# Q7 . Har team ka: average rank, average points nikalo aur analyze karo Kya better rank hamesha better points ka sign hai?
team_avg = df.groupby('Team')[['Rank', 'Points']].mean()
team_avg
```

```python
team_avg['Rank'].corr(team_avg['Points'])
```

```python
# Q8 Kaunsi team ka points trend unstable hai?
df.groupby('Team')['Points'].std().sort_values(ascending=False)

```

```python
# q9 Har year me: Rank aur Points ke beech correlation nikaalo Kya lower rank (1) hamesha higher points deta hai?
df.groupby('Year').apply(
    lambda x: x['Rank'].corr(x['Points'])
)
```

```python
# Q10. Aisi teams identify karo jo Rank 1 par aayi lekin us year highest points nahi the?
max_points = df.groupby('Year')['Points'].transform('max')
max_points
```
