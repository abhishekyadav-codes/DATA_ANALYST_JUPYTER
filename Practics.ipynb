```python
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ABHISHEK YADAV\Downloads\school_marks_dataset.csv")
df['Total']=df[['Maths','Science','English','Hindi']].sum(axis=1)
df['Percentage']=df['Total']/4
df                  

```

```python
def Grade(x):
    if x > 90:
        return ('A')
    elif x >80:
        return ('B')
    elif x >70:
        return ('C')
    else:
        return ('D')

df['Grade']=df['Percentage'].apply(Grade)
df
        
    
```

```python
df['Rank']=df['Percentage'].rank(ascending=False,method='dense').astype(int)
df
# df[df.Total==df['Total'].max()]
```
