```python
df.info()
```

```python
s=df['Revenue']
s
```

```python
df['Revenue']=df['Revenue'].fillna('0')
df
    
```

```python
p=df['Revenue']
p
```

```python
r=[]
for i in p:
    s=""
    for j in i:
        if(j.isdigit()):
            s=s+j
    r.append(s)
r
```

```python
df['Revenue']=r
df
```

```python
df['Revenue']=df['Revenue'].astype('int')
df
```

```python
df.info()
```

```python
s=df['Expenses']
s
```
