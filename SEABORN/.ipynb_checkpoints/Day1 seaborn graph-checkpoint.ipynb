```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
s=sns.load_dataset('tips')
s
```

```python
sns.barplot(x='total_bill',y='tip',data=s)
```

```python
sns.scatterplot(x='total_bill',y='tip',hue='sex',data=s)
 
```

```python
sns.lineplot(x='total_bill',y='tip',hue='sex',data=s)
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',col='day',data=s)
plt.show()
```

```python
sns.relplot(x='total_bill',y='tip',hue='sex',col='time',data=s)
plt.show()
```

```python
v=sns.load_dataset('flights')
v
```

```python
b=v.groupby('month')['passengers'].sum()
```

```python
plt.figure(figsize=(10,5))
plt.plot(b.index,b.values,marker='o')
```

```python
b=v.groupby('year')['passengers'].sum()
plt.figure(figsize=(10,5))
plt.plot(b.index,b.values,marker='o')
```

```python
sns.barplot(x='year',y='passengers',hue='month',data=v)
```
