```python
import pandas as pd
import numpy as np
```

```python
np.random.seed(42)

n = 200

car_names = ["ritz", "sx4", "ciaz", "wagon r", "swift", "verna", "city", "i20", "baleno", "creta"]
fuel_types = ["Petrol", "Diesel","CNG","Electric"]
seller_types = ["Dealer", "Individual"]
transmissions = ["Manual", "Automatic"]

data = {
    "Car_Name": np.random.choice(car_names, n),
    "Year": np.random.randint(2005, 2022, n),
    "Present_Price": np.round(np.random.uniform(3, 20, n), 2),
    "Kms_Driven": np.random.randint(5000, 150000, n),
    "Fuel_Type": np.random.choice(fuel_types, n),
    "Seller_Type": np.random.choice(seller_types, n),
    "Transmission": np.random.choice(transmissions, n),
    "Owner": np.random.randint(0, 3, n)
}

df = pd.DataFrame(data)
# Selling price = Present price ka 50%–90%
df["Selling_Price"] = np.round(df["Present_Price"] * np.random.uniform(0.5, 0.9, n), 2)

# Save file
df.to_csv("car_price_dataset_200_rows.csv", index=False)

print(df.head())
```
