```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df=sns.load_dataset('flights')
df
```

```python
# Q1. Year-wise total passengers ka line plot banao. Har year ka total passengers calculate karo aur trend dekho.year
a=df.groupby('year')['passengers'].sum()
a
plt.figure(figsize=(10,5))
plt.plot(a.index, a.values, marker='o')

plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.title('Year-wise Total Passengers Trend')

plt.show()

```

```python
# Q2. Month-wise average passengers ka bar chart banao.Kaunsa month average me sabse zyada passengers deta hai?
a=df.groupby('month')['passengers'].mean().astype('int')
plt.figure(figsize=(10,5))
plt.bar(a.index, a.values)

plt.xlabel('Year')
plt.ylabel('Total Passengers')
plt.title('Year-wise Total Passengers Trend')

plt.show()
```

```python
# Q3. Passengers ka histogram banao aur distribution observe karo. Data skewed hai ya symmetric? 
 
plt.figure(figsize=(10,6))
plt.hist(df['passengers'], bins=10)
plt.xlabel('passengers')
plt.ylabel('Frequency')
plt.title('Distribution of passenger')
plt.show()
```

```python
#4ï¸ Year vs Passengers simple line plot banao.Overall growth trend dikh raha hai ya nahi?
a=df.groupby('year')['passengers'].sum()
plt.figure(figsize=(10,5))
plt.plot(a.index,a.values, marker='o')
plt.xlabel('year')
plt.ylabel('passengers')
plt.title('graph of year vs passengers')
 
plt.show
```

```python
# Q5 Har year ke liye monthly trend plot karo (multiple lines in one graph).Har year ko alag line se represent karo.
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

df['month'] = pd.Categorical(df['month'],
                             categories=month_order,
                             ordered=True)

df = df.sort_values(['year','month'])
plt.figure(figsize=(12,6))

for year in df['year'].unique():
    year_data = df[df['year'] == year]
    
    plt.plot(
        year_data['month'],
        year_data['passengers'],
        marker='o',
        label=year
    )

plt.xlabel('Month')
plt.ylabel('Passengers')
plt.title('Monthly Trend for Each Year')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

```python
# Q6 Ek hi graph me har month ka long-term trend plot karo. Example: Har saal ka January trend, February trend alag line me.
month_order = ['Jan','Feb','Mar','Apr','May','Jun',
               'Jul','Aug','Sep','Oct','Nov','Dec']

df['month'] = pd.Categorical(df['month'],
                             categories=month_order,
                             ordered=True)
plt.figure(figsize=(12,6))

for month in df['month'].unique():
    month_data = df[df['month'] == month]
    
    plt.plot(
        month_data['year'],
        month_data['passengers'],
        marker='o',
        label=month
    )

plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Long-Term Trend of Each Month')
plt.legend(bbox_to_anchor=(1.05,1))
plt.show()
```

```python
# Q7 Year-wise average passengers ka bar chart banao.Kis year me average growth sabse zyada hai?
a = df.groupby('year')['passengers'].mean()

plt.figure(figsize=(10,5))
plt.bar(a.index, a.values)

plt.xlabel('Year')
plt.ylabel('Average Passengers')
plt.title('Year-wise Average Passengers')

plt.show()
```
