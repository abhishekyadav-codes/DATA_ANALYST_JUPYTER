```python
# Q1. City-wise total crime count ka bar chart banao. Kaun sa city me sabse zyada crimes hue?
a=df['City'].value_counts()
plt.figure(figsize=(10,10))
plt.barh(a.index,a.values)
plt.xlabel('crime count')
plt.ylabel('City')
plt.title('City-wise total crime count ')
```
