```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
df=pd.read_csv(r"C:\Users\ABHISHEK YADAV\OneDrive\Desktop\crime_dataset_india.csv")
df
```

```python
# Q1. City-wise total crime count ka bar chart banao. Kaun sa city me sabse zyada crimes hue?
a=df['City'].value_counts()
plt.figure(figsize=(10,5))
plt.barh(a.index,a.values)
plt.xlabel('crime count')
plt.ylabel('City')
plt.title('City-wise total crime count ')
```

```python
# Q2. Crime Domain (Violent Crime vs Other Crime) ka bar chart banao.Kaunsa domain zyada common hai?

filter_data=df[df['Crime Domain'].isin(['Violent Crime','Other Crime'])]
a=filter_data['Crime Domain'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(a.index, a.values)
plt.xlabel('Crime Domain')
plt.ylabel('Number of Crimes')
plt.title('Crime Domain Distribution')
for i,j in enumerate(a.values):
     plt.text(i,j,str(j),ha='center')
plt.show()




# -------------THIS IS FOR ALL CRIME DOMAIN DATA CRIME ----------------------
# df['Crime Domain'].value_counts()
# plt.figure(figsize=(20,10))
# plt.bar(a.index, a.values)

# plt.xlabel('Crime Domain')
# plt.ylabel('Number of Crimes')
# plt.title('Crime Domain Distribution')
# for i,j in enumerate(a.values):
#     plt.text(i,j,str(j),ha='center')

# plt.show()
```

```python
#️Q3.  Weapon Used ka frequency bar chart banao. Sabse zyada kaunsa weapon use hua?
a=df['Weapon Used'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(a.index,a.values)
plt.xlabel('Weapon Count')
plt.ylabel('Weapon Name')
plt.title('Weapon Used frequency')
for i,j in enumerate(a.values):
    plt.text(i,j,str(j))
plt.show()
    


```

```python
# Q4 .Victim Gender distribution ka pie chart banao.Male vs Female victim ratio kya hai?
a=df['Victim Gender'].value_counts() 
plt.figure(figsize=(6,6))

plt.pie(
    a.values,
    labels=a.index,
    autopct='%1.1f%%',
    startangle=90
)

plt.title('Victim Gender Distribution')
plt.show()
```

```python
# Q5. Victim Age ka histogram banao.Age distribution kaisa hai? Skewed ya symmetric?
plt.figure(figsize=(8,5))

plt.hist(df['Victim Age'], bins=15)

plt.xlabel('Victim Age')
plt.ylabel('Frequency')
plt.title('Victim Age Distribution')

plt.show()
```

```python
# Q6. Case Closed (Yes/No) ka bar chart banao.Kitne percent cases close hue hain?
a=df['Case Closed'].value_counts()
plt.figure(figsize=(10,5))
plt.bar(a.index,a.values)
plt.xlabel('cases yes/no')
plt.ylabel('Frequency')
plt.title('Case closed chart')
plt.show()
```

```python
# Q7. Year-wise crime trend ka line plot banao. Kya crimes time ke saath increase ho rahe hain?
df['Date of Occurrence']=pd.to_datetime(df['Date of Occurrence'])
```

```python
df['Year']=df['Date of Occurrence'].dt.year
df
```

```python
a=df['Year'].value_counts()
a
plt.figure(figsize=(15,5))
plt.plot(a,marker='o')
plt.xlabel('Years')
plt.ylabel('Crime rate ')
plt.title('year wise crime trend')
plt.show()
```

```python
# Q8 Month-wise crime trend plot karo. Kya seasonality exist karti hai?
df['Months']=df['Date of Occurrence'].dt.month_name()
a=df['Months'].value_counts()
plt.figure(figsize=(15,5))
plt.plot(a,marker='o')
plt.xlabel('Years')
plt.ylabel('Crime rate ')
plt.title('Months wise crime trend')
plt.show()



```

```python
# Q9. City-wise violent crimes ka stacked bar chart banao. Har city me violent vs non-violent compare karo.
ct =  df.groupby('City')['Crime Domain'].value_counts
ct
plt.plot(ct,
    
    stacked=True,
    figsize=(12,6)
)

plt.xlabel('City')
plt.ylabel('Number of Crimes')
plt.title('City-wise Violent vs Non-Violent Crimes')

plt.xticks(rotation=45)
plt.show()
```
