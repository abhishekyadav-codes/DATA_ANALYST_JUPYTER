```python
# THE SECOND LARGEST NUMBER OF ARRAY
nums = [10, 25, 8, 40, 15]

largest = nums[0]
second = nums[0]

for i in nums:
    if i > largest:
        second = largest
        largest = i
    elif i > second and i != largest:
        second = i

print("Second Largest:", second)
```

```python
# REVERSE THE ARRAY
arr = [21,54,36,54,25,54]
left = 0
right = len(arr)-1
while left<right:
    arr[left] , arr[right] = arr[right] , arr[left]
    left +=1
    right -=1
print(arr)
```

```python
# CHECK IF ARRAY IS STORED OR NOT 
arr =[1,2,3,4,6,7,8,9]
is_sorted = 'yes'
for i in range(len(arr)-1):
    if arr[i]>arr[i+1]:
        is_sorted='no'
        break
print(is_sorted)
        
```

```python
# REMOVE DUBLICATES OF AN SORTED ARRAY 
arr = [1, 2, 2, 2, 3]

i = 0

for j in range(1, len(arr)):
    if arr[i] != arr[j]:
        i += 1
        arr[i] = arr[j]

print(arr[:i+1])
```

```python
# MOVE ZEROES TO LAST POSITION 
arr=[1,2,6,5,4,0,1,0,2,0]
j=0
for i in range(len(arr)):
    if arr[i]!=0:
        arr[i],arr[j]=arr[j],arr[i]
        j+=1
print(arr)
```

```python
# FIND DUBLICATE NUMBERS
arr = [1, 3, 4, 2, 2]

seen = set()

for i in arr:
    if i in seen:
        print(i)
        break
    else:
        seen.add(i)
```

```python
class Employee:
    company = "TCS"

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show(self):  # Instance
        print(self.name, self.salary)

    @classmethod
    def change_company(cls, new_name):  # Class
        cls.company = new_name

    @staticmethod
    def bonus(salary):  # Static
        return salary * 0.1

```

```python
# polymorphism= same function name but defferent behaviour
# encapsulation=hiding data and controlling access of using object
# inheritance=the class allow to inhert there method to another class
# abtraction= hiding the detils and show essential data 
```
