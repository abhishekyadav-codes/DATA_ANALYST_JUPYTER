```python
# THE SECOND LARGEST NUMBER OF ARRAY
nums = [10, 25, 8, 40, 15]

largest = nums[0]
second = nums[0]

for i in nums:
    if i > largest:
        second = largest
        largest = i
    elif i > second and i != largest:
        second = i

print("Second Largest:", second)
```

```python
# THE LARGEST NUMBER OF ARRAY
nums=[20,25,65,89,21,62,11]
largest_no = nums[0]
for i in nums:
    if i > largest_no:
        largest_no = i
print(largest_no)
    
```

```python
# REVERSE THE ARRAY
arr = [21,54,36,54,25,54]
left = 0
right = len(arr)-1
while left<right:
    arr[left] , arr[right] = arr[right] , arr[left]
    left +=1
    right -=1
print(arr)
```
